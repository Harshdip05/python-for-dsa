```python
def even_or_odd(num):
    """this function finds even or odd"""
    if num % 2 == 0:
        print("the number is even")
    else :
        print("the number is odd")         
```

```python
# call this function
even_or_odd(4)
```

```python
# function with multiple parameters
def add(a,b):
    """ """
    c = a + b
    return c 
result = add(5,6)
print(result)    
print(add(4,5))
```

```python
# default parameters
def greet(name):
    print(f"hello {name}")
greet("Harsh")    
```

```python
def greet(name = "Harshdip"):
    print(f"hello {name}")
greet() 
```

```python
def greet(name = "Harshdip"):
    print(f"hello {name}")
greet("Harsh") 
```

```python
# variable length arguments
# postional and keyword arguments

def print_numbers(*args):
    for numbers in args:
        print(numbers)

print_numbers(1,2,3,4,5,"hi")        
```

```python

def print_numbers(*raja):
    for numbers in raja:
        print(numbers)

print_numbers(1,2,3,4,5,"hi") 
```

```python
# keyword arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="raj",age=12,country="INDIA")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"positional argumnet : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,"raj",name="raja",age=12,country="INDIA")
```

```python
print_details(1,2,3,4,5,"raj",name="raja",age=12,country="INDIA",10)

# postional argument before keyword argument
```
