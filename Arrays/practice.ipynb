### largest number in given array

```python
def get_largest(arr):
    max_arr = arr[0]
    for i in range(len(arr)):
        if arr[i]>max_arr:
            max_arr = arr[i]
    return max_arr        

arr = [12,1,435,5,33]
print(get_largest(arr))
```

```python
def find_largest(arr):
    max_num = float('-inf')
    
    for num in arr:
        if num > max_num:
            max_num = num
    
    return max_num


arr = [4, 7, 2, 9, 5]
print(find_largest(arr))

# Time Complexity == O(n) == We check each element once.
# Space Complexity == O(1) == Only one variable (max_num) is used.
```

### Reverse an Array
Reversing an array means changing the order of elements from end to start

| Method           | Time | Space |
| ---------------- | ---- | ----- |
| New array copy   | O(n) | O(n)  |
| Two pointer swap | O(n) | O(1)  |


```python
def reverse_array(arr):
    new_arr = []

    for i in range(len(arr)-1, -1, -1):
        new_arr.append(arr[i])

    return new_arr


arr = [1,2,3,4,5]
print(reverse_array(arr))
```

Use reverse() when: You want to modify the same list

Use reversed() when: You want a reversed version without changing original data

| Feature               | `reverse()` | `reversed()`                 |
| --------------------- | ----------- | ---------------------------- |
| Changes original list | ✔           | ✘                            |
| Returns new list      | ✘           | ✔ (after converting to list) |
| Type                  | list method | built-in function            |



### Pairs in an array

A pair means two elements taken together from an array.

Example array:
[1, 2, 3]

Possible pairs:
(1,2)
(1,3)
(2,3)

Each pair contains two different elements.

### If array size = n          
### Total pairs: n(n-1)/2

```python
def print_pairs(arr):
    for i in range(len(arr)):
        print("first loop index is ",i)
        for j in range(i+1, len(arr)):
            print(arr[i], arr[j])


arr = [1,2,3,4]
print_pairs(arr)
```
