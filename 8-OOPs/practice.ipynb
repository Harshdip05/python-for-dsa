```python
class Student:
    name = "harsh"
    def __init__(self,name):
        self.name = name
        print("adding new student in database")
    

s1 = Student("karan")
print(s1)
print(s1.name)

s2 = Student("Arjun")
print(s2)
print(s2.name)
```

```python
class Student:
    name = "harsh"
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
        print("adding new student in database")
    

s1 = Student("karan",97)
print(s1)
print(s1.name)
print(s1.marks)

s2 = Student("Arjun",100)
print(s2)
print(s2.name)
print(s1.marks)
```

```python
class Student():
    def __init__(self,name,roll_no,marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

s1 = Student("Harsh",1,90)
print(s1)
print(s1.name,s1.roll_no,s1.marks)

s2 = Student("Karan",2,96)
print(s2)
print(s2.name,s2.roll_no,s2.marks)

s3 = Student("Raj",3,88)
print(s3)
print(s3.name,s3.roll_no,s3.marks)

```
