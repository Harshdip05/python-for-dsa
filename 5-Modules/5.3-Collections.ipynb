```python
from collections import Counter 

data = ["a","b","a","c","d","e","f","e","e","e","e","f","z","a"]
print(data)

count = Counter(data)
print(count)

print(count["a"])
print(count["e"])
# print(count[a])

print(count.most_common())   # returns all elements from highest frequency to lowest
print(count.most_common(1))
print(count.most_common(2))  ## top 2 most common

count.update([1,2])
print(count)
print(data)

count.subtract([2])
print(count)
print(data)

print(dict(count))   ## convert Counter to dictionary

for key,value in count.items():   ## loop through Counter
    print(key,value)
```

```python
from collections import Counter

numbers = [1, 2, 2, 3, 3, 3, 4]

count = Counter(numbers)

print(count)
print(count["1"])   ## "1" is a string, not an integer
print(count[1])     ## 1 is an integer, and it exists in the Counter

print(count[6])
# If a key does not exist, it returns 0 instead of an error.
# In normal dictionaries this would cause KeyError.
```

```python
from collections import Counter
data = [1,2,2,3,3,3]
c = Counter(data)
print(c)


# without counter
data = [1,2,2,3,3,3]
count = {}
for item in data:
    if item in count:      # Does this item already exist as a key in the dictionary count?
        count[item] += 1
    else:
        count[item] = 1
print(count)        
```

```python
# arithmetic with counters

c1 = Counter([1,2,2])
c2 = Counter([2,3])
print(c1+c2)
print(c1-c2)
print(c2-c1)
```

| Feature         | Normal Dictionary 🧾 | Counter ⚡                  |
| --------------- | -------------------- | -------------------------- |
| Counting        | Manual loop          | Automatic                  |
| Missing key     | Error                | Returns 0                  |
| Most common     | Need sorting         | Built-in (`most_common()`) |
| Update counts   | Manual               | `.update()`                |
| Math operations | Not easy             | Supported (`+`, `-`, etc.) |


```python
# min no of balloon
from collections import Counter

text = "balloon"
count = Counter(text)
print(count)
print(min(count["b"],count["a"],count["l"]//2,count["o"]//2,count["n"]))



text = "balloon"
b = a = l = o = n = 0
for ch in text:
    if ch == 'b':
        b += 1
    elif ch == 'a':
        a += 1
    elif ch == 'l':
        l += 1
    elif ch == 'o':
        o += 1
    elif ch == 'n':
        n += 1
print(min(b,a,l//2,o//2,n))        
```
