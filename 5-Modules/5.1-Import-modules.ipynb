##### Importing Modules in Python: Modules and Packages
In Python, modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math

math.sqrt(16)

# You are importing a built-in module that contains mathematical functions.

# Internally, math is: A C-implemented extension module that wraps fast mathematical functions from the system’s C standard library.
# So:
# Python (high-level language)
# ⬇
# math module (written in C)
# ⬇
# C standard library (libm)
# ⬇
# CPU-level math instructions

# That’s why it is very fast.
```

```python
# from package_name import functions

from math import sqrt,pi


print(sqrt(20))
print(sqrt(25))
print(sqrt(36))

print(pi)
```

```python
print(dir(math))
# prints a list of all attributes inside the math module.

print(type(math))
```

```python
# Everything in Python is an object.
# So:
# math → object
# dir() → inspects object attributes
# Attributes → stored in object's namespace

# Internally:
# Python checks something like:
math.__dict__       # Which holds all attributes.
```
